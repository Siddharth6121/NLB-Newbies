```python
import pandas as pd
from transformers import pipeline

# Load the CSV file
df = pd.read_csv('test_without_duplicates.csv')
df1 = sentences = df['Review'].tolist()

# Initialize the sentiment analysis pipeline
classifier = pipeline("sentiment-analysis", model="siebert/sentiment-roberta-large-english")

# Get sentiment labels for the sentences
results = classifier(sentences)
print(results)

# Convert sentiment labels to the specified format
def convert_sentiment(sentiment):
    if sentiment == 'NEGATIVE':
        return [0, 0, 1]
    elif sentiment == 'POSITIVE':
        return [0, 1, 0]
    else:  # Neutral
        return [1, 0, 0]

# Predict sentiment for each sentence and convert to the required format
df['sentiment'] = df['Review'].apply(lambda x: classifier(x)[0]['label'].upper())
df['sentiment label'] = df['sentiment'].apply(convert_sentiment)

# Save the DataFrame with sentiment labels
df.to_csv("sentiment_labeled_test_data.csv", sep='\t', index=False)
```
